```python
import numpy as np # импорт библиотеки
```

```python
list_example = [1, 2, 4, 5, 5, 6]
arr = np.array(list_example) # создание массива из списка
arr
```

```python
zer = np.zeros((2, 2)) # создание массива размером 2х2, состоящего из нулей
zer
```

```python
one = np.ones((2, 2, 3)) # создание массива 2х2х3, состоящего из единиц
one
```

```python
aran = np.arange(2, 7) # создание массива, состоящего из диапазона целых чисел от 2 до 6 включительно
aran
```

```python
lin = np.linspace(0, 1, 101) # создание массива, состоящего из 101 элемента, равномерно распределенных на промежутке от 0 до 1 включительно
lin
```

```python
one.shape # форма трехмерного тензора, состоящего из единиц
```

```python
one.dtype # тип данных массива, состоящего из единиц
```

```python
lin[50] # элемент массива под 50-ым индексом
```

```python
zer[1, 1] # элемент массива по индесом (1, 1)
```

```python
lin[1::2] # срез массива, состоящий из каждго второго элемента массива, начиная с элемента с индексом 1
```

```python
arr1 = np.arange(6).reshape(2, 3) # создание массива целых чисел и изменение его формы
arr2 = np.array([[8, 5, 4], [2, 4, 5]]) # создание массива из двумерного списка
arr1 + arr2 # поэлементное сложение массивов
```

```python
arr1 * arr2  # поэлементное умножение массивов
```

```python
np.mean(arr1) # среднее значение массива
```

```python
np.max(arr1) # максимальное значение массива
```
